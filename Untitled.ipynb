```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
file_path = "cleaned_data.csv"  # Set the correct file path
data = pd.read_csv(file_path)
```

```python
# Set visualization style
sns.set_style("whitegrid")
plt.figure(figsize=(15, 8))
```

```python
# 1. Age distribution of customers (Histogram)
plt.subplot(2, 3, 1)
sns.histplot(data['Age'], bins=20, kde=True, color='blue')
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
```

```python
# 2. Average purchase amount by income level (Bar Chart)
plt.subplot(2, 3, 2)
income_purchase = data.groupby("Income_Level")["Purchase_Amount"].mean().sort_values()
sns.barplot(x=income_purchase.index, y=income_purchase.values, palette="viridis")
plt.title("Average Purchase Amount by Income Level")
plt.xlabel("Income Level")
plt.ylabel("Average Purchase ($)")
```

```python
# 3. Relationship between satisfaction and purchase amount (Scatter Plot)
plt.subplot(2, 3, 3)
sns.scatterplot(x=data['Customer_Satisfaction'], y=data['Purchase_Amount'], alpha=0.6, color='red')
plt.title("Customer Satisfaction vs. Purchase Amount")
plt.xlabel("Satisfaction Level")
plt.ylabel("Purchase Amount ($)")
```

```python
# 4. Payment method usage percentage (Pie Chart)
plt.subplot(2, 3, 4)
payment_counts = data["Payment_Method"].value_counts()
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Payment Method Usage Percentage")
```

```python
# 5. Box Plot for purchase amount distribution by income level
plt.subplot(2, 3, 5)
sns.boxplot(x="Income_Level", y="Purchase_Amount", data=data, palette="coolwarm")
plt.title("Purchase Amount Distribution by Income Level")
plt.xlabel("Income Level")
plt.ylabel("Purchase Amount ($)")
```
