```python
import pandas as pd
import numpy as np
import pickle

from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

# 1. ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
df = pd.read_csv('Deep_clean.csv')

# 2. ÿ•ÿπÿØÿßÿØ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
features = ['Customer_Loyalty_Program_Member', 'Customer_Satisfaction', 'Purchase_Amount', 'Location']
X = df[features].copy()
y = df['Purchase_Amount']

# ÿ™ÿ≠ŸàŸäŸÑ Location ÿ•ŸÑŸâ ÿ£ÿ±ŸÇÿßŸÖ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ Encoding ŸÑŸà ŸÉÿßŸÜÿ™ Categorical
if X['Location'].dtype == 'object':
    X['Location'] = X['Location'].astype('category').cat.codes

# 3. ÿ™ŸÇÿ≥ŸäŸÖ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ TimeSeriesSplit ÿ£Ÿà Train-Test Split ÿ®ÿ≤ŸÖŸÜŸä
split_index = int(len(df) * 0.8)
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

# 4. ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨ (ÿ•ÿ≤ÿßŸÑÿ© Linear Regression)
models = {
    "Random Forest": RandomForestRegressor(n_estimators=100, max_depth=10, min_samples_split=2, random_state=42),
    "XGBoost": XGBRegressor(n_estimators=100, max_depth=10, learning_rate=0.1, random_state=42)
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)

    mae = mean_absolute_error(y_test, predictions)
    rmse = np.sqrt(mean_squared_error(y_test, predictions))  # ÿ™ÿπÿØŸäŸÑ ŸÑÿ≠ÿ≥ÿßÿ® RMSE
    mape = mean_absolute_percentage_error(y_test, predictions)

    results[name] = {
        "MAE": mae,
        "RMSE": rmse,
        "MAPE": mape
    }

    # ÿ≠ŸÅÿ∏ ÿßŸÑŸÖŸàÿØŸäŸÑ
    with open(f"{name.lower().replace(' ', '_')}_model.pkl", "wb") as f:
        pickle.dump(model, f)

# ÿπÿ±ÿ∂ ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨
print(results)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

# ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
df = pd.read_csv('Deep_clean.csv')

# ÿ•ÿπÿØÿßÿØ ÿßŸÑÿÆÿµÿßÿ¶ÿµ ŸàÿßŸÑŸáÿØŸÅ
features = ['Customer_Loyalty_Program_Member', 'Customer_Satisfaction', 'Purchase_Amount', 'Location']
X = df[features].copy()
y = df['Purchase_Amount']

# ÿ™ÿ≠ŸàŸäŸÑ Location ÿ•ŸÑŸâ ÿ£ÿ±ŸÇÿßŸÖ ŸÑŸà ŸÉÿßŸÜÿ™ categorical
if X['Location'].dtype == 'object':
    X['Location'] = X['Location'].astype('category').cat.codes

# ÿ™ŸÇÿ≥ŸäŸÖ ÿßŸÑÿ≤ŸÖŸÜ ÿ•ŸÑŸâ train Ÿà test (80/20)
split_index = int(len(df) * 0.8)
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

# --------- 1. Hyperparameter Tuning using GridSearchCV ---------
rf = RandomForestRegressor(random_state=42)

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid,
                           cv=3, scoring='neg_mean_absolute_error',
                           verbose=2, n_jobs=-1)

grid_search.fit(X_train, y_train)

best_rf_model = grid_search.best_estimator_
print("‚úÖ Best Parameters for Random Forest:")
print(grid_search.best_params_)

# --------- 2. Model Evaluation ---------
predictions = best_rf_model.predict(X_test)

mae = mean_absolute_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))  # ŸÑÿß ÿ™ÿ≥ÿ™ÿÆÿØŸÖ squared=
mape = mean_absolute_percentage_error(y_test, predictions)

print("\nüìä Model Evaluation Metrics:")
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAPE: {mape:.2%}")

# --------- 3. Residual Analysis ---------
residuals = y_test.values - predictions

plt.figure(figsize=(10, 5))
plt.plot(residuals, marker='o', linestyle='-', color='purple')
plt.title("Residuals over Time")
plt.xlabel("Index")
plt.ylabel("Residual (Actual - Predicted)")
plt.axhline(y=0, color='black', linestyle='--')
plt.grid(True)
plt.tight_layout()
plt.show()

print("\nüîç Residuals Summary:")
print(f"Mean of Residuals: {np.mean(residuals):.2f}")
print(f"Std Dev of Residuals: {np.std(residuals):.2f}")
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
import matplotlib.pyplot as plt
import numpy as np

# --------- 1. Hyperparameter Tuning for XGBoost ---------
xgb = XGBRegressor(objective='reg:squarederror', random_state=42)

param_grid_xgb = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 10],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0]
}

grid_search_xgb = GridSearchCV(estimator=xgb,
                                param_grid=param_grid_xgb,
                                cv=3,
                                scoring='neg_mean_absolute_error',
                                verbose=2,
                                n_jobs=-1)

grid_search_xgb.fit(X_train, y_train)

best_xgb_model = grid_search_xgb.best_estimator_
print("‚úÖ Best Parameters for XGBoost:")
print(grid_search_xgb.best_params_)

# --------- 2. Model Evaluation ---------
predictions_xgb = best_xgb_model.predict(X_test)

mae_xgb = mean_absolute_error(y_test, predictions_xgb)
rmse_xgb = np.sqrt(mean_squared_error(y_test, predictions_xgb))
mape_xgb = mean_absolute_percentage_error(y_test, predictions_xgb)

print("\nüìä XGBoost Model Evaluation Metrics:")
print(f"MAE: {mae_xgb:.2f}")
print(f"RMSE: {rmse_xgb:.2f}")
print(f"MAPE: {mape_xgb:.2%}")

# --------- 3. Residual Analysis ---------
residuals_xgb = y_test.values - predictions_xgb

plt.figure(figsize=(10, 5))
plt.plot(residuals_xgb, marker='o', linestyle='-', color='green')
plt.title("XGBoost Residuals over Time")
plt.xlabel("Index")
plt.ylabel("Residual (Actual - Predicted)")
plt.axhline(y=0, color='black', linestyle='--')
plt.grid(True)
plt.tight_layout()
plt.show()

print("\nüîç Residuals Summary (XGBoost):")
print(f"Mean of Residuals: {np.mean(residuals_xgb):.2f}")
print(f"Std Dev of Residuals: {np.std(residuals_xgb):.2f}")
```
